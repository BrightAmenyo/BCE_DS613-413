Stopwords are the most common words in a language that provide little or no additional meaning to a sentence. They are often filtered out of text before processing it for natural language processing (NLP) tasks, as they can add noise and make it harder to identify the important words in a document.
Some examples of English stopwords include:
the
is
in
for
where
when
to
at
These words are so common that they appear in almost every sentence, regardless of the topic. As a result, they do not provide much information about the specific content of a document.
For example, consider the following two sentences:
The cat sat on the mat.
The dog ran in the park.
The stopwords "the" and "in" appear in both sentences, but they do not add any new information. We can remove them without changing the meaning of the sentences:
Cat sat on mat.
Dog ran park.
While these sentences are not grammatically correct, they still convey the same basic meaning as the original sentences.
Removing stopwords can help to improve the performance of NLP models by reducing the amount of noise in the data. It can also make it easier to identify the important words in a document, which can be useful for tasks such as text classification and information retrieval.
In Python, you can use the nltk.corpus.stopwords module to access a list of stopwords for a given language. For example, the following code loads the list of English stopwords:

Lemmatization
Lemmatization is the process of reducing a word to its base or dictionary form, known as the lemma. This is done by considering the word's context and morphological structure. Lemmatization is more sophisticated than stemming, which simply removes the end of a word without regard to its meaning.
Stemming
Stemming is a simpler process than lemmatization. It involves removing the end of a word to obtain a base form. However, stemming does not consider the word's context or morphological structure. This can lead to errors, such as the following:
The word "running" would be stemmed to "run," which is not the base form of the word.
The word "mice" would be stemmed to "mic," which is not a valid word.
Lemmatization
Lemmatization takes into account the word's context and morphological structure to determine its base form. This results in more accurate results than stemming. For example:
The word "running" would be lemmatized to "run."
The word "mice" would be lemmatized to "mouse."
When to Use Lemmatization
Lemmatization is generally preferred over stemming when accuracy is important. This is especially true for tasks such as:
Information retrieval
Text summarization
Machine translation
Natural language processing
How to Perform Lemmatization
Lemmatization can be performed using a variety of tools and libraries. Some popular options include:
The WordNetLemmatizer from the NLTK library
The spaCy library
The Stanford CoreNLP library

```python
from bs4 import BeautifulSoup
import re

def review_to_words(raw_review):
    # 1. Delete HTML
    review_text = BeautifulSoup(raw_review, 'html.parser').get_text()

    # 2. Make a space
    letters_only = re.sub('[^a-zA-Z]', ' ', review_text)

    # 3. Lower letters
    words = letters_only.lower().split()

    # 5. Stopwords
    meaningful_words = [word for word in words if word not in stopwords]

    # 6. Lemmatization
    lemmatizer = WordNetLemmatizer()  # Initialize the lemmatizer
    lemmatize_words = [lemmatizer.lemmatize(word) for word in meaningful_words]
    # 7. space join words
    return ' '.join(lemmatize_words)
```

```python
filtered_df['review_clean'] = filtered_df['review'].apply(review_to_words)
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, svm
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split

# import some data to play with
iris = datasets.load_iris()
X = filtered_df
y = filtered_df.review_clean
class_names = iris.target_names

# Split the data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
y_train = y_train.astype(int)
y_test = y_test.astype(int)

# Run classifier, using a model that is too regularized (C too low) to see
# the impact on the results
classifier = svm.SVC(kernel="linear", C=0.01).fit(X_train, y_train)

np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
titles_options = [
    ("Confusion matrix, without normalization", None),
    ("Normalized confusion matrix", "true"),
]
for title, normalize in titles_options:
    disp = ConfusionMatrixDisplay.from_estimator(
        classifier,
        X_test,
        y_test,
        display_labels=class_names,
        cmap=plt.cm.Blues,
        normalize=normalize,
    )
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```
